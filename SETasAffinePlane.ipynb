```python
def BaseConversion(number, base):
    NewNumber = ''
    if number == 0:
        return '00'
    while number > 0:
        NewDigit = number%base
        NewNumber = str(NewDigit) + NewNumber
        number = int(number/base)
    if len(NewNumber) < 2:
        return str('0' + str(NewNumber))
    return (NewNumber)

Blocks = []

RowBlocks = []
ColumnBlocks = []

for i in range(9):
    Row = []
    for j in range(9):
        Row.append(BaseConversion(i,3) + BaseConversion(j,3))
    RowBlocks.append(Row)
    
for i in range(9):
    Col = []
    for j in range(9):
        Col.append(BaseConversion(j,3) + BaseConversion(i,3))
    ColumnBlocks.append(Col)
    
Blocks.append(RowBlocks)
Blocks.append(ColumnBlocks)

for square in range(len(CompleteSet)):  
    SquareBlock = []
    for number in range(9):
        NumberList = []
        for i in range(9):
            for j in range(9):
                if CompleteSet[square][i][j] == number:
                    NumberList.append(BaseConversion(i,3) + BaseConversion(j,3))
        SquareBlock.append(NumberList)
    Blocks.append(SquareBlock)
    
for i in Blocks:
    for j in i:
        print('[%s]' % ', '.join(map(str, j)))
    print('')
```
